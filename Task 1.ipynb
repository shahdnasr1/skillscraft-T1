```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

# Create a bar chart
sns.countplot(x='continent', data=df, palette='viridis')

# Add labels and a title
plt.title('Number of Countries by Continent')  # Title of the chart
plt.xlabel('Continent')  # Label for the x-axis
plt.ylabel('Count')  # Label for the y-axis

# Show the chart
plt.show()

```

```python
# Create a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['2023 population'], bins=20, kde=True, color='blue')

# Add labels and title
plt.title('Distribution of 2023 Population')
plt.xlabel('Population')
plt.ylabel('Frequency')

# Show the chart
plt.show()
```

### Top 10 countries by Population in 2023

```python
df.groupby("continent")["2023 population"].mean().sort_values(ascending =False).tail(2)
```

```python
Top_10_countries = df.sort_values(by="2023 population",ascending= False).head(10)
```

```python
Top_10_countries
```

```python
sns.barplot(x="2023 population",y="country",data =Top_10_countries)
plt.title("Top 10 countries by Population in 2023")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
```
