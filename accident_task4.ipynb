## Handling missing Values

```python
#checking missing values
df.isna().sum()
```

```python
df.head()
```

```python
#storing categorical columns name to a new variable
categorical =[i for i in df.columns if df[i].dtype=='O']
print("the categorical variables are",categorical)
```

```python
#for categorical values we can replace the null values with the Mode of it
for i in categorical:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

# Seperating Independent and dependent

```python
x= df3.drop(['Accident_severity'],axis=1)
x.shape
```

```python
x.head()
```

```python
y = df3.iloc[:,2]
y.head()
```

```python
#plotting count plot using seaborn
sns.countplot(x=y,palette='muted')
```

## oversampling

```python
#importing SMOTE
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
xo,yo = oversample.fit_resample(x,y)
```

```python
#checking the oversampling output
y1= pd.DataFrame(yo)
y1.value_counts()
```

```python
sns.countplot(x = yo,palette='muted')
```

# spliting the data

```python
#converting data to training data and testing dara
from sklearn.model_selection import train_test_split
#spliting 70% of the data to training data and 30% of data to testing data
x_train,x_test,y_train,y_test=train_test_split(xo,yo,test_size=0.30,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

## KNN MODEL CREATION

## prediction

```python
#KNN model alg
from sklearn.neighbors import KNeighborsClassifier
model_KNN =KNeighborsClassifier(n_neighbors=5)
model_KNN.fit(x_train,y_train)
```

```python
y_pred =model_KNN.predict(x_test)
```

```python
y_pred
```

# checking Accurancy,classification Report,confusion Matrix

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, ConfusionMatrixDisplay
```

#### classification Report

```python
report_KNN=classification_report(y_test,y_pred)
print(report_KNN)
```

```python
accurancy_KNN =accuracy_score(y_test,y_pred)
print(accurancy_KNN)
```

#### confusion matrix

```python
matrix_KNN =confusion_matrix(y_test,y_pred)
print(matrix_KNN,'\n')
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
